```
# BFS

def bfs_shortest_path(graph_set, inicio, meta):
  
    print('Ciudad de origen: ',inicio)
    print('Ciudad destino: ',meta)
    print('\n')
    
    # hacer un seguimiento de los nodos explorados
    explored = []
    # realizar un seguimiento de todos los caminos que deben comprobarse
    queue = [[inicio]]
    
    # retornar la ruta si el inicio es el objetivo
    if inicio == meta:
        return "La ciudad destino es igual a la ciudad de origen"
 
    # mantenernos en bucle hasta que se hayan verificado todos los caminos posibles
    while queue:
        # Pop el primer camino de la cola
        path = queue.pop(0)
        
        # obtener el último nodo de la ruta
        node = path[-1]
        print('Ruta en la cola: ',path, ' - Último nodo de la ruta ',node)
        if node not in explored:
            print(f'Nodo no explorado {node} ')
            neighbours = graph_set[node]
            # Ir a través de todos los nodos vecinos, construir una nueva ruta y empujarlo en la cola
            print('Nodos vecinos de : ',node,' son: ',neighbours)
            for neighbour in neighbours:
                new_path = list(path)                
                new_path.append(neighbour)
                queue.append(new_path)
                print('Nuevas Rutas', new_path)
                
                # retornar ruta si el vecino es el objetivo
                if neighbour == meta:
                  return 'La ruta más factible escogida de acuerdo al algoritmo es: ', new_path
            print('UPS!! No se encontro la ciudad destino')
            print('Nuevas rutas agregada a la cola: ',queue)
            print('-------------------------------------------------------------------------------------------------------------')

            # marca el nodo como explorado
            explored.append(node)
        else:
            print(f'Ya ha sido explorado nodo: {node}')
 
    # En caso de que no haya ruta entre los 2 nodos.
    return "Lo siento, pero no existe una ruta de conexión..."
```

```
#Definimos una funcion con Parametros

def dfs_path(graph, start, goal):
  
    print('Nodo inicio: ',start)
    print ('Nodo Objetivo/Meta: ',goal)
    print('\n')
  
    # Se define en el stack con el Nodo de Inicio = start
    stack = [[start]]
    
    # Declaramos un While que lo que va a hacer es a recorrer todo el arbol(grafo), mostrando los caminos posibles para llegar al 
    # Objetivo Final Aplicando recursividad como es el caso de este ejercicio en donde utiliza la operacion POP.
    while stack:
        path = stack.pop()
        node = path[-1]
        
        print('Se muestra la Ruta en el Stack: ',path, 'Último nodo en la ruta: ',node)
        
        #Ciclo For 
        for next in graph[node] - set(path):
            
            # If a correct path is founded, then return the path with the generator
            # else write a new path and follow iterating.
            
            #Condicional que establece que si encuentra el objetivo muestre la ruta que ha recorrido
            #Caso Contrario crea una nueva ruta hasta llegar al objetivo final, verificando si se reviso con anterioridad los nodos
            #de otros caminos
            
            
            if next == goal:
                print('El Camino escogido fue: ', path + [next], ' - Objetivo Final: ',goal)
                print('------------------------------------------------------------------------------------------------------------------------')
                print('Recursividad - Ruta alternativa')
                path + [next]
               

            else:
                #Agrega un nodo/item al final de la lista
                
                stack.append(path + [next])
```

```
import networkx as nx
import matplotlib.pyplot as plt

maxdepth = 0
explored = []
fringe = []

def IDDFS(root):
    
    global explored
    global fringe
    global maxdepth
    
    for depth in range(200):
        print("\033[1mITERATION: \033[0m" + str(depth))
        explored = []
        fringe = []
        maxdepth = depth
        found = DLS(root, depth, [])
        if found is not None:
            return found
    
def DLS(node, depth, parents):
    
    if node in parents:
        #Node Single
        return
   
    print("\033[94mExplored: \033[0m {: <6} ------------------> Fringe: {: <7} \nExploring: {: <13} Depth: {: <2}".format(str(parents), 
        str(fringe), str(node.name), str(maxdepth - depth)))
    
    parents.append(node)

    if node not in explored:
        explored.append(node)
    if fringe:
        fringe.pop(0)

    if depth == 0 and node is goal:
        return [node]
    
    if depth > 0:
        for child in reversed(node.children):
            #List Node Neighbours
            if child not in parents:
                fringe.insert(0, child)
        for child in node.children:
            copy = parents[:] 
            found = DLS(child, depth-1, copy)
            if found is not None:
                found.insert(0, node)
                #Ruta
                return found
    return

class Node:
    def __init__(self, name, children):
        self.name = name
        self.children = children

    def __init__(self, name):
        self.name = name
        self.children = []

    def __repr__(self):
        return self.name

    def add(self, node):
        self.children.append(node)

    def adds(self, *nodes):
        for node in nodes:
            self.add(node)
```

```
from collections import deque

class Graph:
    def __init__(self, directed=True):
        self.edges = {}
        self.directed = directed

    def add_edge(self, node1, node2, __reversed=False):
        try: neighbors = self.edges[node1]
        except KeyError: neighbors = set()
        neighbors.add(node2)
        self.edges[node1] = neighbors
        if not self.directed and not __reversed: self.add_edge(node2, node1, True)

    def neighbors(self, node):
        try: return self.edges[node]
        except KeyError: return []

    def depth_limited_search(self, start, goal, limit=-1):
        print('Depth limit =', limit)
        found, fringe, visited, came_from = False, deque([(0, start)]), set([start]), {start: None}
        print('{:13s} | {}'.format('Expand Node', 'Fringe'))
        print('--------------------')
        print('{:13s} | {}'.format('-', start))
        while not found and len(fringe):
            depth, current = fringe.pop()
            print('{:13s}'.format(current), end=' | ')
            if current == goal: found = True; break
            if limit == -1 or depth < limit:
                for node in self.neighbors(current):
                    if node not in visited:
                        visited.add(node); fringe.append((depth + 1, node))
                        came_from[node] = current
            print(', '.join([n for _, n in fringe]))
        if found: print(); return came_from
        else: print('No path from {} to {}'.format(start, goal))

    @staticmethod
    def print_path(came_from, goal):
        parent = came_from[goal]
        if parent:
            Graph.print_path(came_from, parent)
        else: print(goal, end='');return
        print(' =>', goal, end='')

    def __str__(self):
        return str(self.edges)


graph = Graph(directed=False)
graph.add_edge('AMBATO', 'RIOBAMBA')
graph.add_edge('AMBATO', 'LATACUNGA')
graph.add_edge('AMBATO', 'GUARANDA')
graph.add_edge('BABAHOYO', 'PORTOVIEJO')
graph.add_edge('BABAHOYO', 'DURAN')
graph.add_edge('BABAHOYO', 'QUEVEDO')
graph.add_edge('BABAHOYO', 'MILAGRO')
graph.add_edge('BABAHOYO', 'DAULE')
graph.add_edge('CAYAMBE', 'QUITO')
graph.add_edge('CAYAMBE', 'IBARRA')
graph.add_edge('CAYAMBE', 'QUININDE')
graph.add_edge('CAYAMBE', 'OTAVALO')
graph.add_edge('CAYAMBE', 'RUMIÑAHUI')
graph.add_edge('CHONE', 'PORTOVIEJO')
graph.add_edge('CHONE', 'MANTA')
graph.add_edge('CHONE', 'QUEVEDO')
graph.add_edge('CHONE', 'EL_CARMEN')
graph.add_edge('CHONE', 'MONTECRISTI')
graph.add_edge('CUENCA', 'DURAN')
graph.add_edge('CUENCA', 'MACHALA')
graph.add_edge('CUENCA', 'LOJA')
graph.add_edge('CUENCA', 'RIOBAMBA')
graph.add_edge('CUENCA', 'MILAGRO')
graph.add_edge('CUENCA', 'LAGO_AGRIO')
graph.add_edge('DAULE', 'GUAYAQUIL')
graph.add_edge('DAULE', 'PORTOVIEJO')
graph.add_edge('DAULE', 'QUEVEDO')
graph.add_edge('DAULE', 'MILAGRO')
graph.add_edge('DAULE', 'BABAHOYO')
graph.add_edge('DAULE', 'MONTECRISTI')
graph.add_edge('DAULE', 'SAMBORONDON')
graph.add_edge('DURAN', 'GUAYAQUIL')
graph.add_edge('DURAN', 'SAMBORONDON')
graph.add_edge('DURAN', 'CUENCA')
graph.add_edge('DURAN', 'MACHALA')
graph.add_edge('DURAN', 'RIOBAMBA')
graph.add_edge('DURAN', 'MILAGRO')
graph.add_edge('DURAN', 'BABAHOYO')
graph.add_edge('EL_CARMEN', 'SANTO_DOMINGO')
graph.add_edge('EL_CARMEN', 'PORTOVIEJO')
graph.add_edge('EL_CARMEN', 'MANTA')
graph.add_edge('EL_CARMEN', 'ESMERALDA')
graph.add_edge('EL_CARMEN', 'CHONE')
graph.add_edge('EL_CARMEN', 'QUININDE')
graph.add_edge('ESMERALDAS', 'PORTOVIEJO')
graph.add_edge('ESMERALDAS', 'MANTA')
graph.add_edge('ESMERALDAS', 'IBARRA')
graph.add_edge('ESMERALDAS', 'LATACUNGA')
graph.add_edge('ESMERALDAS', 'QUININDE')
graph.add_edge('ESMERALDAS', 'LAGO_AGRIO')
graph.add_edge('ESMERALDAS', 'MONTECRISTI')
graph.add_edge('ESMERALDAS', 'TULCAN')
graph.add_edge('GUARANDA', 'AMBATO')
graph.add_edge('GUARANDA', 'RIOBAMBA')
graph.add_edge('GUARANDA', 'BABAHOYO')
graph.add_edge('GUAYAQUIL', 'PORTOVIEJO')
graph.add_edge('GUAYAQUIL', 'DURAN')
graph.add_edge('GUAYAQUIL', 'SANTA_ELENA')
graph.add_edge('GUAYAQUIL', 'DAULE')
graph.add_edge('GUAYAQUIL', 'MONTECRISTI')
graph.add_edge('GUAYAQUIL', 'SAMBORONDON')
graph.add_edge('IBARRA', 'ESMERALDAS')
graph.add_edge('IBARRA', 'OTAVALO')
graph.add_edge('IBARRA', 'LAGO_AGRIO')
graph.add_edge('IBARRA', 'CAYAMBE')
graph.add_edge('IBARRA', 'TULCAN')
graph.add_edge('LAGO_AGRIO', 'ESMERALDAS')
graph.add_edge('LAGO_AGRIO', 'IBARRA')
graph.add_edge('LAGO_AGRIO', 'QUITO')
graph.add_edge('LAGO_AGRIO', 'CUENCA')
graph.add_edge('LAGO_AGRIO', 'TULCAN')
graph.add_edge('LATACUNGA', 'QUITO')
graph.add_edge('LATACUNGA', 'MEJIA')
graph.add_edge('LATACUNGA', 'SANTO_DOMINGO')
graph.add_edge('LATACUNGA', 'QUEVEDO')
graph.add_edge('LATACUNGA', 'AMBATO')
graph.add_edge('LIBERTAD', 'PORTOVIEJO')
graph.add_edge('LIBERTAD', 'SANTA_ELENA')
graph.add_edge('LIBERTAD', 'MONTECRISITI')
graph.add_edge('LIBERTAD', 'MANTA')
graph.add_edge('LOJA', 'CUENCA')
graph.add_edge('LOJA', 'RIOBAMBA')
graph.add_edge('MACHALA', 'DURAN')
graph.add_edge('MACHALA', 'RIOBAMBA')
graph.add_edge('MACHALA', 'CUENCA')
graph.add_edge('MACHALA', 'MILAGRO')
graph.add_edge('MANTA', 'SANTA_ELENA')
graph.add_edge('MANTA', 'CHONE')
graph.add_edge('MANTA', 'LIBERTAD')
graph.add_edge('MANTA', 'MONTECRISTI')
graph.add_edge('MEJIA', 'QUITO')
graph.add_edge('MEJIA', 'SANTO_DOMINGO')
graph.add_edge('MEJIA', 'LATACUNGA')
graph.add_edge('MILAGRO', 'CUENCA')
graph.add_edge('MILAGRO', 'DURAN')
graph.add_edge('MILAGRO', 'BABAHOYO')
graph.add_edge('MILAGRO', 'MACHALA')
graph.add_edge('MONTECRISTI', 'GUAYAQUIL')
graph.add_edge('MONTECRISTI', 'PORTOVIEJO')
graph.add_edge('MONTECRISTI', 'MANTA')
graph.add_edge('MONTECRISTI', 'SANTA_ELENA')
graph.add_edge('MONTECRISTI', 'DAULE')
graph.add_edge('MONTECRISTI', 'CHONE')
graph.add_edge('MONTECRISTI', 'SAMBORONDON')
graph.add_edge('OTAVALO', 'IBARRA')
graph.add_edge('OTAVALO', 'CAYAMBE')
graph.add_edge('OTAVALO', 'QUITO')
graph.add_edge('OTAVALO', 'QUININDE')
graph.add_edge('OTAVALO', 'RUMIÑAHUI')
graph.add_edge('PORTOVIEJO', 'GUAYAQUIL')
graph.add_edge('PORTOVIEJO', 'MANTA')
graph.add_edge('PORTOVIEJO', 'QUEVEDO')
graph.add_edge('PORTOVIEJO', 'CHONE')
graph.add_edge('PORTOVIEJO', 'MONTECRISTI')
graph.add_edge('PORTOVIEJO', 'LIBERTAD')
graph.add_edge('QUEVEDO', 'SANTO_DOMINGO')
graph.add_edge('QUEVEDO', 'PORTOVIEJO')
graph.add_edge('QUEVEDO', 'RIOBAMBA')
graph.add_edge('QUEVEDO', 'LATACUNGA')
graph.add_edge('QUEVEDO', 'BABAHOYO')
graph.add_edge('QUEVEDO', 'DAULE')
graph.add_edge('QUEVEDO', 'CHONE')
graph.add_edge('QUEVEDO', 'GUARANDA')
graph.add_edge('QUININDE', 'QUITO')
graph.add_edge('QUININDE', 'SANTO_DOMINGO')
graph.add_edge('QUININDE', 'ESMERALDAS')
graph.add_edge('QUININDE', 'OTAVALO')
graph.add_edge('QUININDE', 'EL_CARMEN')
graph.add_edge('QUININDE', 'CAYAMBE')
graph.add_edge('QUININDE', 'MEJIA')
graph.add_edge('QUITO', 'SANTO_DOMINGO')
graph.add_edge('QUITO', 'LATACUNGA')
graph.add_edge('QUITO', 'QUININDE')
graph.add_edge('QUITO', 'OTAVALO')
graph.add_edge('QUITO', 'LAGO_AGRIO')
graph.add_edge('QUITO', 'LIBERTAD')
graph.add_edge('QUITO', 'CAYAMBE')
graph.add_edge('QUITO', 'MEJIA')
graph.add_edge('QUITO', 'RUMIÑAHUI')
graph.add_edge('RIOBAMBA', 'CUENCA')
graph.add_edge('RIOBAMBA', 'AMBATO')
graph.add_edge('RIOBAMBA', 'PORTOVIEJO')
graph.add_edge('RIOBAMBA', 'DURAN')
graph.add_edge('RIOBAMBA', 'MACHALA')
graph.add_edge('RIOBAMBA', 'LOJA')
graph.add_edge('RIOBAMBA', 'QUEVEDO')
graph.add_edge('RIOBAMBA', 'MILAGRO')
graph.add_edge('RIOBAMBA', 'BABAHOYO')
graph.add_edge('RIOBAMBA', 'GUARANDA')
graph.add_edge('RUMIÑAHUI', 'QUITO')
graph.add_edge('RUMIÑAHUI', 'SANTO_DOMINGO')
graph.add_edge('RUMIÑAHUI', 'QUININDE')
graph.add_edge('RUMIÑAHUI', 'OTAVALO')
graph.add_edge('RUMIÑAHUI', 'LAGO_AGRIO')
graph.add_edge('RUMIÑAHUI', 'CAYAMBE')
graph.add_edge('SAMBORONDON', 'DAULE')
graph.add_edge('SAMBORONDON', 'DURAN')
graph.add_edge('SAMBORONDON', 'GUAYAQUIL')
graph.add_edge('SANTA_ELENA', 'GUAYAQUIL')
graph.add_edge('SANTA_ELENA', 'PORTOVIEJO')
graph.add_edge('SANTA_ELENA', 'LIBERTAD')
graph.add_edge('SANTA_ELENA', 'MANTA')
graph.add_edge('SANTO_DOMINGO', 'QUITO')
graph.add_edge('SANTO_DOMINGO', 'QUEVEDO')
graph.add_edge('SANTO_DOMINGO', 'LATACUNGA')
graph.add_edge('SANTO_DOMINGO', 'QUININDE')
graph.add_edge('SANTO_DOMINGO', 'EL_CARMEN')
graph.add_edge('SANTO_DOMINGO', 'MEJIA')
graph.add_edge('TULCAN', 'IBARRA')
graph.add_edge('TULCAN', 'ESMERALDAS')
graph.add_edge('TULCAN', 'LAGO_AGRIO')


start, goal, l = 'GUAYAQUIL', 'QUITO', 7
traced_path = graph.depth_limited_search(start, goal, l)
if (traced_path): print('Path:', end=' '); Graph.print_path(traced_path, goal);print()
```

Luego tenemos un arco que tiene un nodo from_node y un nodo to_node y un costo (no negativo):

```
class Arc(object):
    """Un arco tiene un nodo from_node y un nodo to_node y un costo (no-negativo)"""
    def __init__(self, from_node, to_node, cost=1, action=None):
        assert cost >= 0, ("Costo no puede ser negativo para"+
                           str(from_node)+"->"+str(to_node)+", cost: "+str(cost))
        self.from_node = from_node
        self.to_node = to_node
        self.action = action
        self.cost=cost

    def __repr__(self):
        """Retorna un  string de la representación de un arco"""
        if self.action:
            return str(self.from_node)+" --"+str(self.action)+"--> "+str(self.to_node)
        else:
            return str(self.from_node)+" --> "+str(self.to_node)
            
```

En la clase de representación explicita del grafo de búsqueda un problema de búsqueda consiste de:
*  una lista o set de nodos: nodes
*	una lista o set de arcos: arcs
*	un nodo inicial (start):  start=None
*	una lista o set de nodos objetivos: goals=set()
* un diccionario que mapea cada nodo en su valor heurístico: hmap={}

Se define un problema de búsqueda, necesitamos definir el nodo de inicio, el predicado objetivo, la función de los vecinos y la función heurística. Lo siguiente se usa para la implementación de búsqueda (DFS) primero en profundidad a continuación 

```
class Search_problem_from_explicit_graph(Search_problem):
    """Un problema de búsqueda consiste de:
    * una lista o set de nodos: nodes
    * una lista o set de arcos: arcs
    * un nodo inicial (start):  start=None
    * una lista o set de nodos objetivos: goals=set()
    * un diccionario que mapea cada nodo en su valor heurístico: hmap={}
    """
  

    def __init__(self, nodes, arcs, start=None, goals=set(), hmap={}):
        self.neighs = {}
        self.nodes = nodes
               

        for node in nodes:
            self.neighs[node]=[]
        self.arcs = arcs
        for arc in arcs:
            self.neighs[arc.from_node].append(arc)
        self.start = start
        self.goals = goals
        self.hmap = hmap

    def start_node(self):
        """retorna el nodo inicial (start)"""
        return self.start
    
    def is_goal(self,node):
        """es True so nodo es un objetivo (goal)"""
        return node in self.goals

    def neighbors(self,node):
        """retorna los vecinos (neighbors) de un nodo"""
        return self.neighs[node]

    def heuristic(self,node):
        """Dado el valor heurístico de un nodo n.
        Retorna 0 si no sobrescribir en el hmap."""
        if node in self.hmap:
            return self.hmap[node]
        else:
            return 0
        
    def __repr__(self):
        """retorna un string de la representación del problema de búsqueda"""
        res=""
        for arc in self.arcs:
            res += str(arc)+".  "
        return res

    def neighbor_nodes(self,node):
        """retorna un iterador sobre los vecinos (neighbors) de un nodo"""
        return (path.to_node for path in self.neighs[node])
```

La cual un buscador devolvera una ruta desde el nodo de inicio a un nodo objetivo. Se utiliza una representacion alternativa en terminos de una estructura de datos recursiva que puede compartir subpartes. Un camino es:
* un nodo (que representa una ruta de longitud 0) o
*	una ruta, inicial y un arco, donde el nodo desde el arco es el nodo en el final de la inicial.

Estos casos se distinguen en el siguiente código al tener arc = None si la ruta tiene una longitud de 0, en cuyo caso el nodo inicial (start) es el nodo de la ruta. 

```
class Path(object):
    """Una ruta (path) es ya sea un nodo o una trayectoria seguida por un arco"""
    
    def __init__(self,initial,arc=None):
        """inicial es un nodo (en cuyo caso el arco es None) o
        una ruta (path) (en cuyo caso arco es un objeto de tipo arco)"""
        self.initial = initial
        self.arc=arc
        if arc is None:
            self.cost=0
        else:
            self.cost = initial.cost + arc.cost

    def end(self):
        """retorna el nodo en el fin de la ruta (path)"""
        if self.arc is None:
            return self.initial
        else:
            return self.arc.to_node

    def nodes(self):
        """enumera los nodos para la ruta.
        Esto comienza al final y enumera los nodos en la ruta hacia atrás."""
        current = self
        while current.arc is not None:
            yield current.arc.to_node
            current = current.initial
        yield current.initial

    def initial_nodes(self):
        """enumera los nodos para la ruta antes del nodo final.
        Esto comienza al final y enumera los nodos en la ruta hacia atrás."""
        if self.arc is not None:
            for nd in self.initial.nodes(): yield nd     # could be "yield from"
        
    def __repr__(self):
        """retorna un string de representación de una ruta (path)"""
        if self.arc is None:
            return str(self.initial)
        elif self.arc.action:
            return (str(self.initial)+"\n   --"+str(self.arc.action)
                    +"--> "+str(self.arc.to_node))
        else:
            return str(self.initial)+" --> "+str(self.arc.to_node)
```

Se construye un objeto Searcher para el problema y luego pedir repetidamente la siguiente ruta utilizando la búsqueda. Si no hay más caminos, se devuelve None

```
class Displayable(object):
    """Class that uses 'display'.
    The amount of detail is controlled by max_display_level
    """
    max_display_level = 1   # can be overridden in subclasses

    def display(self,level, *args, **nargs):
        """print the arguments if level is less than or equal to the
        current max_display_level.
        level is an integer.
        the other arguments are whatever arguments print can take.
        """
        if level <= self.max_display_level:
            print(*args, **nargs)  ##if error you are using Python2 not Python3
def visualize(func):
    """A decorator for algorithms that do interactive visualization.
    Ignored here.
    """
    return func
```

```
class Searcher(Displayable):
    """retorna un buscador (searcher) para un problema.
    Las Rutas (Paths) pueden ser encontradas por llamadas repetidas a search().
    Esto hace primero la búsqueda de profundidad a menos que se anule
    """
    def __init__(self, problem):
        """crea un buscador (searcher) desde un problema
        """
        self.problem = problem
        self.initialize_frontier()
        self.num_expanded = 0
        self.add_to_frontier(Path(problem.start_node()))
        super().__init__()

    def initialize_frontier(self):
        self.frontier = []
        
    def empty_frontier(self):
        return self.frontier == []
        
    def add_to_frontier(self,path):
        self.frontier.append(path)
        
    @visualize
    def search(self):
        """retorna (next) path - (siguiente) ruta - desde el nodo inicial (start) de un problema
        a un nodo objetivo (goal). 
        Retorna None si no existe ruta (path).
        """
        while not self.empty_frontier():
          
            path = self.frontier.pop()
            
            self.display(2, "Expandiendo:",path,"(costo:",path.cost,")")
            self.num_expanded += 1
            
            if self.problem.is_goal(path.end()):    # solución encontrada
                self.display(1, self.num_expanded, "caminos se han ampliado y",
                            len(self.frontier), "rutas (paths) permanecen en la frontera")
                self.solution = path   # almacena la solución encontrada
                return path
            else:
                neighs = self.problem.neighbors(path.end())
                self.display(3,"Los vecinos (Neighbors) son", neighs)
                
                for arc in reversed(neighs):
                    self.add_to_frontier(Path(path,arc))
                self.display(3,"Frontera:",self.frontier)
                
        self.display(1,"No (más) soluciones. Total de",
                     self.num_expanded,"rutas (paths) expandidas.")
```

En la clase frontera consiste de una cola (queue) de prioridas (priority) (heap), fronterapq, de triples (value, index, path), dónde:
* 	value es el valor que queremos minimizar (e.g., costo de path + h).
*  index es un índice único para cada elemento
* path es la ruta en la cola (queue)

Tenga en cuenta que la cola de prioridad siempre devuelve el elemento más pequeño

```
import heapq        

#from searchProblem import Path

class FrontierPQ(object):
    """Una frontera consiste de una cola (queue) de prioridas (priority) (heap), fronterapq, de
       triples (value, index, path), dónde
    * value es el valor que queremos minimizar (e.g., costo de path + h).
    * index es un índice único para cada elemento
    * path es la ruta en la cola (queue)
    Tenga en cuenta que la cola de prioridad siempre devuelve el elemento más pequeño.
    """

    def __init__(self):
        """construye la frontera, inicialmente una cola de prioridad vacía
        """
        self.frontier_index = 0  # el número de elementos que se han añadido a la frontera.
        self.frontierpq = []  # la cola de prioridad de la frontera

    def empty(self):
        """es True si la cola de priority está vacía (empty)"""
        return self.frontierpq == []

    def add(self, path, value):
        """añade una ruta (path) a la cola priority
        value es el valor a ser minimizado"""
        self.frontier_index += 1    # obtener un nuevo índice único
        heapq.heappush(self.frontierpq,(value, -self.frontier_index, path))

    def pop(self):
        """retorna y elimina el camino de la frontera con valor mínimo.
        """
        (_,_,path) = heapq.heappop(self.frontierpq)
        return path 

    def count(self,val):
        """retorna el número de elementos de la frontera con value=val"""
        return sum(1 for e in self.frontierpq if e[0]==val)

    def __repr__(self):
        """string de representación de la frontier"""
        return str([(n,c,str(p)) for (n,c,p) in self.frontierpq])
    
    def __len__(self):
        """longitud (length) de la frontier"""
        return len(self.frontierpq)

    def __iter__(self):
        """iterar a través de a través de los caminos en el frontier"""
        for (_,_,path) in self.frontierpq:
            yield path
```

Para una búsqueda A∗, la frontera se implementa utilizando la clase FrontierPQ.

```
class AStarSearcher(Searcher):
    """retorna un buscador searcher para un problema.
    Las rutas se pueden encontrar llamando repetidamente a search().
    """

    def __init__(self, problem):
        super().__init__(problem)

    def initialize_frontier(self):
        self.frontier = FrontierPQ()

    def empty_frontier(self):
        return self.frontier.empty()

    def add_to_frontier(self,path):
        """agregar ruta (path) a la frontera con el costo apropiado"""
        value = path.cost + self.problem.heuristic(path.end())
        print('Costo del camino',path.cost,' + Costo de la heurística de: ',path.end(),' es: ',self.problem.heuristic(path.end()),' --> ',value)
        self.frontier.add(path, value)
```

```
graph_letra={
    'A': set(['B','C','E']),
    'B': set(['A','D','E']),
    'C': set(['A','F','G']),
    'D': set(['B','E']),
    'E': set(['A','B','D']),
    'F': set(['C']),
    'G': set(['C']),
}

def bfs_path(graph_letra, inicio, meta):
  
    print('Ciudad de origen: ',inicio)
    print('Ciudad destino: ',meta)
    print('\n')
    
    # hacer un seguimiento de los nodos explorados
    explored = []
    # realizar un seguimiento de todos los caminos que deben comprobarse
    queue = [[inicio]]
    
    # retornar la ruta si el inicio es el objetivo
    if inicio == meta:
        return "La ciudad destino es igual a la ciudad de origen"
 
    # mantenernos en bucle hasta que se hayan verificado todos los caminos posibles
    while queue:
        # Pop el primer camino de la cola
        path = queue.pop(0)
        
        # obtener el último nodo de la ruta
        node = path[-1]
        print('Ruta en la cola: ',path, ' - Último nodo de la ruta ',node)
        if node not in explored:
            print(f'Nodo no explorado {node} ')
            neighbours = graph_letra[node]
            # Ir a través de todos los nodos vecinos, construir una nueva ruta y empujarlo en la cola
            print('Nodos vecinos de : ',node,' son: ',neighbours)
            for neighbour in neighbours:
                new_path = list(path)                
                new_path.append(neighbour)
                queue.append(new_path)
                print('Nuevas Rutas', new_path)
                
                # retornar ruta si el vecino es el objetivo
                if neighbour == meta:
                  return 'La ruta más factible escogida de acuerdo al algoritmo es: ', new_path
            print('UPS!! No se encontro la ciudad destino')
            print('Nuevas rutas agregada a la cola: ',queue)
            print('-------------------------------------------------------------------------------------------------------------')

            # marca el nodo como explorado
            explored.append(node)
        else:
            print(f'Ya ha sido explorado nodo: {node}')
 
    # En caso de que no haya ruta entre los 2 nodos.
    return "Lo siento, pero no existe una ruta de conexión..."
```

###DFS

```
graph_letra={
    'A': set(['B','C','E']),
    'B': set(['A','D','E']),
    'C': set(['A','F','G']),
    'D': set(['B','E']),
    'E': set(['A','B','D']),
    'F': set(['C']),
    'G': set(['C']),
}

def dfs_path(graph, start, goal):
  
    print('Nodo inicio: ',start)
    print ('Nodo Objetivo/Meta: ',goal)
    print('\n')
  
    # Se define en el stack con el Nodo de Inicio = start
    stack = [[start]]
    
    # Declaramos un While que lo que va a hacer es a recorrer todo el arbol(grafo), mostrando los caminos posibles para llegar al 
    # Objetivo Final Aplicando recursividad como es el caso de este ejercicio en donde utiliza la operacion POP.
    while stack:
        path = stack.pop()
        node = path[-1]
        
        print('Se muestra la Ruta en el Stack: ',path, 'Último nodo en la ruta: ',node)
        
        #Ciclo For 
        for next in graph[node] - set(path):
            
            # If a correct path is founded, then return the path with the generator
            # else write a new path and follow iterating.
            
            #Condicional que establece que si encuentra el objetivo muestre la ruta que ha recorrido
            #Caso Contrario crea una nueva ruta hasta llegar al objetivo final, verificando si se reviso con anterioridad los nodos
            #de otros caminos
            
            
            if next == goal:
                print('El Camino escogido fue: ', path + [next], ' - Objetivo Final: ',goal)
                print('------------------------------------------------------------------------------------------------------------------------')
                print('Recursividad - Ruta alternativa')
                path + [next]
               

            else:
                #Agrega un nodo/item al final de la lista
                
                stack.append(path + [next])
```

```
from collections import deque

class Graph:
    def __init__(self, directed=True):
        self.edges = {}
        self.directed = directed

    def add_edge(self, node1, node2, __reversed=False):
        try: neighbors = self.edges[node1]
        except KeyError: neighbors = set()
        neighbors.add(node2)
        self.edges[node1] = neighbors
        if not self.directed and not __reversed: self.add_edge(node2, node1, True)

    def neighbors(self, node):
        try: return self.edges[node]
        except KeyError: return []

    def depth_limited_search(self, start, goal, limit=-1):
        print('Depth limit =', limit)
        found, fringe, visited, came_from = False, deque([(0, start)]), set([start]), {start: None}
        print('{:13s} | {}'.format('Expand Node', 'Fringe'))
        print('--------------------')
        print('{:13s} | {}'.format('-', start))
        while not found and len(fringe):
            depth, current = fringe.pop()
            print('{:13s}'.format(current), end=' | ')
            if current == goal: found = True; break
            if limit == -1 or depth < limit:
                for node in self.neighbors(current):
                    if node not in visited:
                        visited.add(node); fringe.append((depth + 1, node))
                        came_from[node] = current
            print(', '.join([n for _, n in fringe]))
        if found: print(); return came_from
        else: print('No path from {} to {}'.format(start, goal))

    @staticmethod
    def print_path(came_from, goal):
        parent = came_from[goal]
        if parent:
            Graph.print_path(came_from, parent)
        else: print(goal, end='');return
        print(' =>', goal, end='')

    def __str__(self):
        return str(self.edges)


graph = Graph(directed=False)
graph.add_edge('A', 'B')
graph.add_edge('A', 'E')
graph.add_edge('A', 'C')
graph.add_edge('B', 'A')
graph.add_edge('B', 'D')
graph.add_edge('B', 'E')
graph.add_edge('C', 'F')
graph.add_edge('C', 'G')
graph.add_edge('D', 'B')
graph.add_edge('D', 'E')
graph.add_edge('E', 'A')
graph.add_edge('E', 'B')
graph.add_edge('E', 'D')
graph.add_edge('F', 'C')
graph.add_edge('G', 'C')

start, goal, l = 'D', 'F', 4
traced_path = graph.depth_limited_search(start, goal, l)
if (traced_path): print('Path:', end=' '); Graph.print_path(traced_path, goal);print()
```

```
from queue import heappop, heappush
from math import inf

class Graph:
    def __init__(self, directed=True):
        self.edges = {}
        self.directed = directed

    def add_edge(self, node1, node2, cost = 1, __reversed=False):
        try: neighbors = self.edges[node1]
        except KeyError: neighbors = {}
        neighbors[node2] = cost
        self.edges[node1] = neighbors
        if not self.directed and not __reversed: self.add_edge(node2, node1, cost, True)

    def neighbors(self, node):
        try: return self.edges[node]
        except KeyError: return []

    def cost(self, node1, node2):
        try: return self.edges[node1][node2]
        except: return inf


    def uniform_cost_search(self, start, goal):
        found, fringe, visited, came_from, cost_so_far = False, [(0, start)], set([start]), {start: None}, {start: 0}
        print('{:11s} | {}'.format('Expand Node', 'Fringe'))
        print('--------------------')
        print('{:11s} | {}'.format('-', str((0, start))))
        while not found and len(fringe):
            _, current = heappop(fringe)
            print('{:11s}'.format(current), end=' | ')
            if current == goal: found = True; break
            for node in self.neighbors(current):
                new_cost = cost_so_far[current] + self.cost(current, node)
                if node not in visited or cost_so_far[node] > new_cost:
                    visited.add(node); came_from[node] = current; cost_so_far[node] = new_cost
                    heappush(fringe, (new_cost, node))
            print(', '.join([str(n) for n in fringe]))
        if found: print(); return came_from, cost_so_far[goal]
        else: print('No path from {} to {}'.format(start, goal)); return None, inf

    @staticmethod
    def print_path(came_from, goal):
        parent = came_from[goal]
        if parent:
            Graph.print_path(came_from, parent)
        else: print(goal, end='');return
        print(' =>', goal, end='')


    def __str__(self):
        return str(self.edges)
    
graph = Graph(directed=True)
graph.add_edge('A', 'B', 4)
graph.add_edge('A', 'E', 1)
graph.add_edge('A', 'C', 3)
graph.add_edge('B', 'A', 4)
graph.add_edge('B', 'D', 3)
graph.add_edge('B', 'E', 8)
graph.add_edge('C', 'F', 4)
graph.add_edge('C', 'G', 4)
graph.add_edge('D', 'B', 3)
graph.add_edge('D', 'E', 5)
graph.add_edge('E', 'A', 1)
graph.add_edge('E', 'B', 8)
graph.add_edge('E', 'D', 5)
graph.add_edge('F', 'C', 4)
graph.add_edge('G', 'C', 4)
start, goal = 'D', 'F'

#PRUEBA D a F

traced_path, cost = graph.uniform_cost_search(start, goal)
if (traced_path): print('Path:', end=' '); Graph.print_path(traced_path, goal); print('\nCost:', cost)
```
